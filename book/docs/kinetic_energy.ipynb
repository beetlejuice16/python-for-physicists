```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rcParams
plt.style.use('seaborn-darkgrid')
```

```python
%matplotlib inline
rcParams['font.size'] = '20'
rcParams['figure.figsize'] = (15,12)
```

```python
v = np.arange(0, 11)
v_squared = v**2
m = 5
```

```python
def E_kin(v:np.ndarray, m:float) -> np.ndarray:
    return 0.5 * m * v**2
```

```python
def E_pot(h, m):
    g = 9.81
    return m*g*h
```

```python
def eom1(v_0, a, t):
    s = v_0*t + 0.5 * a * t**2
    return s
```

```python
def eom2(v_0, a, s):
    v_squared = v_0**2 + 2*a*s
    return v_squared
```

```python
Total_E = np.ones(len(v)) * E_kin(v[-1], 5)
```

```python
t = np.arange(0, 11)
```

```python
v = np.sqrt(eom2(0, 9.81, eom1(0, 9.81, t)))
```

```python
kin_energy = E_kin(v, 5)
```

```python
pot_energy = E_pot(eom1(0, 9.81, t)[::-1], 5)
```

```python
fig, ax = plt.subplots(figsize=(14,10))
ax.plot(v, kin_energy, 'o-')
ax.plot(v, pot_energy, 'o-')
ax.plot(v, kin_energy + pot_energy)
ax.set_title('Kinetic Energy vs Velocity')
ax.set_xlabel('v [m/s]')
ax.set_ylabel('$E_{kin}$ [J]')
```
